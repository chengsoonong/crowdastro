# ATLAS Gold Standard

This notebook generates a gold standard of ATLAS objects using Norris classifications.

```python
import astropy.io.ascii as asc, numpy
```

```python
table = asc.read('one-table-to-rule-them-all.tbl')
resolved = table[table['Component S (Franzen)'] / table['Component Sp (Franzen)'] > 1]
resolved_rgz = resolved[numpy.array([bool(i) for i in resolved['Component Zooniverse ID (RGZ)']])]
resolved_rgz_norris = resolved_rgz[numpy.array([bool(i) for i in resolved_rgz['Source # (Norris)']])]
resolved_rgz_norris
```

```python
numpy.random.seed(0)
indices = numpy.random.choice(list(range(len(resolved_rgz_norris))), replace=False, size=20)
```

```python
resolved_rgz_norris[indices]
```

```python
zids = [row['Component Zooniverse ID (RGZ)'] for row in resolved_rgz_norris[indices]]
assert len(set(zids)) == 20
```

```python
with open('atlas_gold_standard.txt', 'w') as f:
    for zid in zids:
        f.write('{}\n'.format(zid))
```
